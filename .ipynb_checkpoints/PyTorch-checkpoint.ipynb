<h1>Autograd - automatyczne różniczkowanie</h1>

```python
x = torch.tensor(2.0, requires_grad=True)

y = x**2 + 3*x + 1

y.backward()

print(x.grad)  # pochodna
```

<h1>Budowa prostego modelu</h1>

```python
import torch.nn as nn

class SimpleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model = SimpleModel()
print(model)
```

<H1>Funkcja straty i optymalizacja</H1>

```python
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

<h1>Proces treningu</h1>

```python
x_train = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)
y_train = 2 * x_train + 1

for epoch in range(100):
    predictions = model(x_train)
    loss = criterion(predictions, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

<h1>Wizualizacja wyników</h1>

```python
import matplotlib.pyplot as plt

predicted = model(x_train).detach().numpy()

plt.scatter(x_train.numpy(), y_train.numpy(), label="Dane")
plt.plot(x_train.numpy(), predicted, color='red', label="Model")
plt.legend()
plt.show()
```

<h1>Dataset i DataLoader</h1>

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self):
        self.x = torch.linspace(-1, 1, 100)
        self.y = 2 * self.x + 1

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

dataset = CustomDataset()
loader = DataLoader(dataset, batch_size=10, shuffle=True)
```

<h1>Model dla obrazów</h1>

```python
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 3)
        self.fc1 = nn.Linear(32*26*26, 10)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        return x
```
